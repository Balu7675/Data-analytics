```python
import os
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials

# Set Spotify API credentials
os.environ['SPOTIPY_CLIENT_ID'] = 'ec8c5b9abc734b3f956c3022ae115676'
os.environ['SPOTIPY_CLIENT_SECRET'] = '95b6803716c243ebbb4026bd6c2478c2'

# SS Thaman Spotify URI
thaman_uri = 'spotify:artist:24YbYIa6wcYPLomzRKBSnj'

# Authenticate and fetch data
spotify = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials())
results = spotify.artist_top_tracks(thaman_uri)

# Extract track info
track_data = []
for track in results['tracks'][:10]:
    track_name = track['name']
    preview_url = track['preview_url']
    cover_art = track['album']['images'][0]['url'] if track['album']['images'] else 'Not available'
    
    track_data.append({
        'Track Name': track_name,
        'Preview URL': preview_url,
        'Cover Art URL': cover_art
    })

# Create DataFrame
df_tracks = pd.DataFrame(track_data)
print(df_tracks)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

class SpotifyAnalysis:
    def __init__(self, client_id, client_secret):
        auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
        self.spotify = spotipy.Spotify(auth_manager=auth_manager)

    def get_followers(self, artist_uris):
        data = []

        for uri in artist_uris:
            try:
                artist = self.spotify.artist(uri)
                data.append({
                    'Name': artist['name'],
                    'Followers': artist['followers']['total'],
                    'Genres': ', '.join(artist['genres']),
                    'Popularity': artist['popularity']
                })
            except Exception as e:
                print(f"Error fetching {uri}: {e}")

        return pd.DataFrame(data)
     
# ====== Usage ======
client_id = 'ec8c5b9abc734b3f956c3022ae115676'
client_secret = '95b6803716c243ebbb4026bd6c2478c2'

sa = SpotifyAnalysis(client_id, client_secret)

artist_uris = ['spotify:artist:5sSzCxHtgL82pYDvx2QyEU']  # Devi Sri Prasad
df_followers = sa.get_followers(artist_uris)

print(df_followers)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

# Initialize Spotipy client
client_id = "ec8c5b9abc734b3f956c3022ae115676"
client_secret = "95b6803716c243ebbb4026bd6c2478c2"

auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager)

# Get album data
album_uri = 'spotify:album:19wxNqNBGMba438PzfG4xm'
album_data = sp.album(album_uri)

# Extract track info
tracks = album_data['tracks']['items']

# Convert to DataFrame
track_list = []

for track in tracks:
    track_info = {
        'Track Name': track['name'],
        'Track Number': track['track_number'],
        'Duration (ms)': track['duration_ms'],
        'Explicit': track['explicit'],
        'Spotify URL': track['external_urls']['spotify']
    }
    track_list.append(track_info)

df = pd.DataFrame(track_list)

# Display the DataFrame
print(df)
```
